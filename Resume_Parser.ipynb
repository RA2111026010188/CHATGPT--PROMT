```python
import fitz  # PyMuPDF
import json

# Define a function to extract text from PDF
def extract_text_from_first_page(pdf_path):
    # Open the PDF file
    doc = fitz.open(pdf_path)

    # Extract text from the first page
    page_text = doc[0].get_text()

    # Close the document
    doc.close()

    return page_text

# Function to parse the extracted text
def parse_resume(text):
    lines = text.split('\n')
    # Print lines for debugging
    for i, line in enumerate(lines):
        print(f"{i}: {line}")

    resume_dict = {}

    personal_details = {
        "name": lines[0] if len(lines) > 0 else "",
        "degree": lines[1] if len(lines) > 1 else "",
        "phone": lines[2].split(': ')[1] if len(lines) > 2 and ': ' in lines[2] else "",
        "email": lines[3].split(': ')[1] if len(lines) > 3 and ': ' in lines[3] else "",
        "current_address": lines[4] if len(lines) > 4 else ""
    }

    skills = lines[6].replace('Skills: ', '').split(', ') if len(lines) > 6 else []

    brief_summary = lines[8] if len(lines) > 8 else ""

    education = []
    if len(lines) > 11:
        for i in range(11, min(19, len(lines)), 3):
            education.append({
                "year": lines[i] if i < len(lines) else "",
                "institution": lines[i + 1].split('|')[0].strip() if i + 1 < len(lines) else "",
                "degree": lines[i + 1].split('|')[1].strip() if i + 1 < len(lines) and '|' in lines[i + 1] else "",
                "cgpa_or_percentage": lines[i + 2] if i + 2 < len(lines) else ""
            })

    internships = []
    if len(lines) > 21:
        internships.append({
            "duration": lines[21] if len(lines) > 21 else "",
            "company": lines[22] if len(lines) > 22 else "",
            "role": lines[23] if len(lines) > 23 else "",
            "description": lines[24] if len(lines) > 24 else ""
        })

    projects = []
    if len(lines) > 28:
        projects.append({
            "duration": lines[28] if len(lines) > 28 else "",
            "title": lines[29] if len(lines) > 29 else "",
            "skills": lines[30].split(', ') if len(lines) > 30 else [],
            "link": lines[31].split(': ')[1] if len(lines) > 31 and ': ' in lines[31] else "",
            "description": [lines[32], lines[33], lines[34], lines[35], lines[36]] if len(lines) > 36 else []
        })
        projects.append({
            "duration": lines[39] if len(lines) > 39 else "",
            "title": lines[40] if len(lines) > 40 else "",
            "mentor": lines[41].split('|')[0].split(': ')[1] if len(lines) > 41 and ': ' in lines[41] else "",
            "team_size": int(lines[41].split('|')[1].split(': ')[1]) if len(lines) > 41 and '|' in lines[41] else 0,
            "skills": lines[42].split(', ') if len(lines) > 42 else [],
            "description": [lines[43], lines[44], lines[45], lines[46], lines[47]] if len(lines) > 47 else []
        })

    publications = []
    if len(lines) > 50:
        publications.append({
            "date": lines[50] if len(lines) > 50 else "",
            "title": lines[51] if len(lines) > 51 else "",
            "mentor": lines[52].split(': ')[1] if len(lines) > 52 and ': ' in lines[52] else "",
            "description": lines[53] if len(lines) > 53 else ""
        })

    achievements = lines[56:59] if len(lines) > 59 else []

    extracurricular_activities = lines[61:63] if len(lines) > 63 else []

    personal_interests = lines[65] if len(lines) > 65 else ""

    personal_details_extra = {
        "gender": lines[67].split(': ')[1] if len(lines) > 67 and ': ' in lines[67] else "",
        "marital_status": lines[68].split(': ')[1] if len(lines) > 68 and ': ' in lines[68] else "",
        "current_address": lines[69].split(': ')[1] if len(lines) > 69 and ': ' in lines[69] else "",
        "date_of_birth": lines[70].split(': ')[1] if len(lines) > 70 and ': ' in lines[70] else "",
        "known_languages": lines[71].split(': ')[1].split(', ') if len(lines) > 71 and ': ' in lines[71] else [],
        "permanent_address": lines[72].split(': ')[1] if len(lines) > 72 and ': ' in lines[72] else ""
    }

    resume_dict['personal_details'] = personal_details
    resume_dict['skills'] = skills
    resume_dict['brief_summary'] = brief_summary
    resume_dict['education'] = education
    resume_dict['internships'] = internships
    resume_dict['projects'] = projects
    resume_dict['publications'] = publications
    resume_dict['achievements'] = achievements
    resume_dict['extracurricular_activities'] = extracurricular_activities
    resume_dict['personal_interests'] = personal_interests
    resume_dict['personal_details_extra'] = personal_details_extra

    return resume_dict

# Extract text from the uploaded resume PDF
pdf_path = '/content/Pod resume[1].pdf'
text = extract_text_from_first_page(pdf_path)

# Parse the text into JSON
resume_json = parse_resume(text)

# Print the JSON
print(json.dumps(resume_json, indent=2))
```
